# **Analysis of Renewable Energy Sector in Europe**

This project investigates long-term trends in Europe’s energy landscape, with a particular focus on the growth of renewable energy sources. It uses annual data from 1985 to 2024 and examines changes in primary energy production, imports and exports, gross available energy, and final energy consumption.
All datasets are publicly available through Our World in Data, Eurostat, and the Energy Institute.

### Goals
The objective of this project is to strengthen my skills in Python, Jupyter Notebook, and data handling with CSV files by working through a real-world dataset and answering analytical questions along the way.
I chose to focus on renewable energy in Europe because it is a field I am particularly interested in and one that plays a crucial role in global sustainability efforts.

This analysis aims to answer the following questions:
* What are the main energy sources in Europe?
* Where does renewable energy stand today in terms of production and consumption?
* How do European countries compare in their total electricity consumption?
* How much of Europe’s energy comes from low-carbon sources?
* Which renewable energy sources are growing the fastest in terms of expansion and investment?


### Highlights
